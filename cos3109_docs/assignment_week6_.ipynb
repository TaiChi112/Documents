**conditional probability**

$P (\mathrm{x} = x; \mathrm{y} = y) = P(x) P(y)$ denotes the probability that $\mathrm{x} = x$ and $\mathrm{y} = y$


มีค่าซ้ำกันหลายชุด แสดงว่าต้อง นับความถี่ของคู่ (x, y) แล้วนำมาหารด้วยจำนวนทั้งหมด เพื่อหาความน่าจะเป็น

$$\color{orange}{\forall x \in \mathrm{x}, P(\mathrm{x} = x) = \displaystyle\sum_{y} P(\mathrm{x} = x, \mathrm{y} = y) = \displaystyle\sum_{y} P( x) P(y)) \tag{1}}$$

$$\color{red}{P(\mathrm{x} = x \ | \ \mathrm{y} = y) = \frac{P(\mathrm{x} = x, \mathrm{y} = y)}{P(\mathrm{y} = y)} \tag{3a}}$$

$$\color{blue}{P(\mathrm{x} = x \ | \ \mathrm{y} = y) = \frac{P( x \cap y)}{P(y)} \tag{3b}}$$


\begin{align*}
2x - 5y &=  8 \\
3x + 9y &=  -12
\end{align*}


\begin{align*}

 &=  0.987
\end{align*}
